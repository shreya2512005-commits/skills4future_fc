```python
class Bank:
    def _init_(self,balance):
        #self.amount = amount
        self.balance = balance
    #I want deposit
    def deposit(self,amount):
        #balance+amount
        self.balance = self.balance + amount
        #if amount is neg add if else statement
    #i want to withdraw
    def withdraw(self,amount):
        #pass
        #balance-amount
        if amount>self.balance:
            print('Insufficient fund!')
        else:
            self.balance= self.balance-amount
    def balance(self):
        return self.balance
```
